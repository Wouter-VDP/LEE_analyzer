## Definitions

```python
# Dictionary translating the category field in the root tree
categories={1: 'Cosmic',
            2: 'e CC',
            3: 'mu CC',
            4: 'NC',
            5: 'Dirt',
            6: 'Data',
            7: 'Mixed',
            0: 'Other'}

# Fiducial volume borders in x,y,z:
fid_arr= [[10,10],[20,20],[10,50]]

# Minimum reconstructable energies:
mass_p= 0.93827
mass_e= 0.00511
min_e = 0.02+mass_e # 20MeV
min_p = 0.04+mass_p # 40MeV

# Dictionary of ROOT files
filedict = OrderedDict([('Nue_cosmics',"Input/nue_cosmic.root")])

# List of fields in the ROOT tree you want to exclude
ignorelist = ['reconstruced_energy']
```

## Functions

```python
# Return true if the point is in the TPC with a tolerance.
def CheckBorderTPC(x,y,z,tolerance=0):
    detectorx   =256.35     # In cm
    detectory   =116.5      # Symmetric around 0     
    detectorz   =1036.8
    d=tolerance # border tolerance
    if (0+d) < x < (detectorx-d):
            if (-detectory+d)< y < (detectory-d):
                    if (0+d) < z < (detectorz-d):
                        return True
    return False

def CheckBorderFid(x,y,z,arr):
    detectorx   =256.35     # In cm
    detectory   =116.5      # Symmetric around 0     
    detectorz   =1036.8
    if (0+arr[0][0]) < x < (detectorx-arr[0][1]):
            if (-detectory+arr[1][0])< y < (detectory-arr[1][1]):
                    if (0+arr[2][0]) < z < (detectorz-arr[2][1]):
                        return True
    return False


# Return the angle between two numpy vectors.
def anglevec(v1,v2):
    v1_u = v1 / np.linalg.norm(v1)
    v2_u = v2 / np.linalg.norm(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))


# Verifies if the event passed the flash precuts.
def flash_precut(row):
    #Flash precuts
    t_start = 3.2
    t_end   = t_start+1.6
    min_PE  = 50

    for time,PE in zip(row['flash_time'],row['flash_PE']):
        if time>t_start and time<t_end and PE>min_PE:
            return True
    return False

# Sum reconstructed energy
def CalcRecoE(row):
    reco_energy = [0,0,0,0]
    total_nhits = [0,0,0]
    
    for tr_hits,tr_dedx,nhits in zip(row['track_energy_hits'],row['track_energy_dedx'],row["track_nhits"]):
        reco_energy[:3]+=tr_hits  
        reco_energy[-1]+=tr_dedx 
        total_nhits+=nhits
        
    for shower,nhits in zip(row['shower_energy'],row["shower_nhits"]):
        reco_energy[:3]+=shower
        reco_energy[-1]+=shower[2] #4th entry uses plane 2 hits for showers and dedx for tracks
        total_nhits+=nhits
        
    return return pd.Series({"reconstructed_energy": reco_energy, "total_nhits": total_nhits})

# Definition 
def true_thresholds(row):
    passed_e=False
    passed_p=False
    for pdg,E in zip(row["nu_daughters_pdg"],row["nu_daughters_E"]):
        if pdg==11 and E>min_e:
            passed_e=True
        if pdg==2212 and E>min_p:
            passed_p=True
    return passed_e and passed_p

# Formatting
def sciNot(x):
    x=float(x)
    return "{:.1f}".format(x)

```

```python
# Load root files into a dataframe or dictionary of dataframes.
def loadData(filedict,                  # List of input files, dictionary
             ignorelist,                # List of fields (columns) to exclude from the dataframe  
             maxn=4300,                 # Maximum number of events (rows) in the final dataframe 
             concat=False,              # In case of multiple input tags, merge dataframes?
             contained=True,            # Is the true neutrino interaction in a defined volume?
             passed=False,              # Did the event pass the selection stage?
             flashpassed=False,         # Was there still a candidate passing the optical cuts?
             truecontains=[11,2212],    # Require the true neutrino interaction to have the following daughters.
             LEE_denom=False ):         # 1 proton of at least 40MeV, 1 electron of at least 20MeV, true vertex inside: x[10,10],y[20,20],z[10,50]
    
    global categories
    global fid_arr
    dfdict={}
    
    for key,fname in filedict.items():
        dftemp=read_root(fname,"robertoana/pandoratree",ignore=ignorelist)
        
        print ('\n\nDataframe loaded from:',fname)
        print ('Dataframe', key, 'has', len(dftemp.index), 'entries.')
        
        if contained:
            dftemp=dftemp[dftemp.apply(lambda x: CheckBorderTPC(x['true_vx'],x['true_vy'],x['true_vz']), axis=1)]
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries with true vertex in TPC.')
            
        for pdg in truecontains:
            dftemp = dftemp[dftemp["nu_daughters_pdg"].apply(lambda x: pdg in x)]
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries with a pdg',pdg,'particle.')
            
        if LEE_denom:
            dftemp=dftemp[dftemp.apply(lambda x: CheckBorderFid(x['true_vx'],x['true_vy'],x['true_vz'],fid_arr), axis=1)]
            #dftemp.drop(['true_vx','true_vy','true_vz'],axis=1,inplace=True)
            dftemp = dftemp[dftemp[["nu_daughters_pdg","nu_daughters_E"]].apply(true_thresholds, axis=1)]
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries with the LEE signal definition.')
        
        dftemp = dftemp[dftemp.apply(flash_precut, axis=1)]
        print ('Dataframe', key, 'has', len(dftemp.index), 'entries with a valid flash.')
        
        dftemp["category"]=dftemp["category"].map(categories)
        dftemp["flash_config"]=dftemp.apply(CalcRecoE, axis=1)
        
        if(maxn>0 and maxn<len(dftemp.index)):
            dftemp=dftemp.sample(n=maxn)
            print ('Final dataframe', key, 'has', len(dftemp.index), 'entries.') 
            
        if flashpassed:
            dftemp = dftemp[ dftemp['flash_passed'].apply(lambda x: not np.all(x==-1))]
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries passing the optical selection.')
            
        if passed:
            dftemp=dftemp[dftemp['passed']==1]
            total = len(dftemp.index)
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries which passed selection.')
            close = len(dftemp[dftemp['distance']<5].index)
            print ('Closeness purity :', close,'/', total,'(',sciNot(close/total*100),'%)','is within 5cm from true_sce vertex.')
            
            
    dftemp.drop(['reconstructed_energy'],axis=1,inplace=True) #because there was a bug
    dftemp[['reconstructed_energy','total_nhits']]=dftemp[["track_energy_hits","shower_energy",'track_energy_dedx',"shower_nhits","track_nhits"]].apply(CalcRecoE,axis=1)
    
    dftemp["flash_config"]=key
    dfdict[key]=dftemp
        
    if concat:
        return pd.concat(dfdict.values(),ignore_index=True,copy=False)
    else:
        return dfdict
```

## Load dataframe and manipulation

```python
not_pickled=False
if not_pickled:
    df_flash = loadData(filedict,
                        ignorelist,
                        maxn        =2000000,
                        concat      =False,
                        contained   =True,
                        passed      =True,
                        flashpassed =True,
                        truecontains=[11,2212],
                        LEE_denom   =True)

    setting = 'Nue_cosmics'
    df = df_flash[setting] #.drop_duplicates(subset='event',keep=False)#
    df.to_pickle('Input/nue_cosmic_passed_LEEdef.pckl')
else:
    df = pd.read_pickle('Input/nue_cosmic_passed_LEEdef.pckl')
    print('Dataframe loaded from Pickle has '+str(len(df.index))+' entries.' )
```

```python
df_energy = df[['category','distance',"shower_nhits","track_nhits",
                'shower_energy','matched_showers_energy','matched_showers',
                'track_energy_dedx','track_energy_hits','matched_tracks_energy','matched_tracks',
                'reconstructed_energy','nu_E']]
```

## Plots & Analysis

```python
# Proton Track Energy Reconstruction

plt.style.use('default')
fig,ax=plt.subplots(ncols=2,nrows=1,figsize=(5*gr,5),sharey=False)
plt.suptitle(r'Energy reconstruction for proton tracks')
N=7*5
maxhit_plane=[0,0,0]

if 1:
    
    arr_true = []
    
    arr_dedx = []
    arr_pl0  = []
    arr_pl1  = []
    arr_pl2  = []
    arr_plmax= []
    
    for index,row in df[df['distance']<5].iterrows():
        
        for e_hit,e_dedx,n_hit,pdg,e_true in zip(row['track_energy_hits'],row['track_energy_dedx'],row['track_nhits'],row["matched_tracks"],row['matched_tracks_energy']):
            if pdg==2212:
                arr_true.append(e_true-mass_p)
                arr_dedx.append(e_dedx)
                arr_pl0.append(e_hit[0])
                arr_pl1.append(e_hit[1])
                arr_pl2.append(e_hit[2])
                max_pl=np.argmax(n_hit)
                arr_plmax.append(e_hit[max_pl])
                maxhit_plane[max_pl]+=1

absolute = [arr_true,arr_dedx,arr_pl0,arr_pl1,arr_pl2,arr_plmax][::-1]
labels = ['True E',r'd$E$/d$x$','hits, plane 0','hits, plane 1','hits, plane 2','hits, plane max'][::-1]
relative = np.divide(np.array(absolute[:-1]),np.array(arr_true))
ax[0].hist(absolute,normed=1, bins=np.logspace(np.log10(0.01),np.log10(2.0), N), histtype='step', stacked=False, fill=False,label=labels)     
ax[0].set_xscale("log")
ax[0].set_xlabel('Kinetic energy [GeV]')
ax[0].set_ylabel('Normalised')
ax[0].legend()

ax[1].hist(relative.T,normed=1, bins=N, histtype='step', fill=False,range=[0,1.6],label=labels[:-1])         
ax[1].set_xlabel('Reconstructed kinetic energy / True kinetic energy')
ax[1].set_ylabel('Normalised')
ax[1].legend()
ax[1].set_ylim(0,2.9)

plt.tight_layout()
fig.subplots_adjust(top=0.9)
plt.savefig('track_E.pdf')

print(np.array(maxhit_plane)/sum(maxhit_plane)*100)
```

```python
# Electron Shower Energy Reconstruction

plt.style.use('default')
fig,ax=plt.subplots(ncols=2,nrows=1,figsize=(5*gr,5),sharey=False)
plt.suptitle(r'Energy reconstruction for electron showers')
N=7*5
maxhit_plane=[0,0,0]

if 1:
    
    arr_true = []
    
    arr_pl0  = []
    arr_pl1  = []
    arr_pl2  = []
    arr_plmax= []
    
    for index,row in df[df['distance']<5].iterrows():
        if 11 not in row["matched_showers"]:
            continue
        e_i = np.argmax(np.array(row["matched_showers"])==11)
        arr_true.append(row['matched_showers_energy'][e_i]-mass_e)
        
        temp_pl    =[0,0,0]
        temp_nhits =[0,0,0]
        
        for e_hit,n_hit,pdg in zip(row['shower_energy'],row['shower_nhits'],row["matched_showers"]):
            if pdg==11:
                temp_pl+=e_hit
                temp_nhits+=n_hit
                
        arr_pl0.append(temp_pl[0])
        arr_pl1.append(temp_pl[1])
        arr_pl2.append(temp_pl[2])
        arr_plmax.append(temp_pl[np.argmax(temp_nhits)])
        maxhit_plane[np.argmax(temp_nhits)]+=1
        
absolute = [arr_true,arr_pl0,arr_pl1,arr_pl2,arr_plmax][::-1]
labels = ['True E','hits, plane 0','hits, plane 1','hits, plane 2','hits, plane max'][::-1]
ax[0].hist(absolute,normed=1, bins=np.logspace(np.log10(0.01),np.log10(2.0), N), histtype='step', stacked=False, fill=False,label=labels)     
ax[0].set_xscale("log")
ax[0].set_xlabel('Kinetic energy [GeV]')
ax[0].set_ylabel('Normalised')
ax[0].legend(loc=3)

relative = np.divide(np.array(absolute[:-1]),np.array(arr_true))
ax[1].hist(relative.T,normed=1, bins=N, histtype='step', fill=False,range=[0,1.85],label=labels[:-1])         
ax[1].set_xlabel('Reconstructed kinetic energy / True kinetic energy')
ax[1].set_ylabel('Normalised')
ax[1].legend()
#ax[1].set_ylim(0,2.9)

plt.tight_layout()
fig.subplots_adjust(top=0.9)
plt.savefig('shower_E.pdf')
print(np.array(maxhit_plane)/sum(maxhit_plane)*100)
```
