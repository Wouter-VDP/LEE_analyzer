## Functions

```python
# Return true if the point is in the TPC with a tolerance.
def CheckBorderTPC(x,y,z,tolerance=0):
    detectorx   =256.35     # In cm
    detectory   =116.5      # Symmetric around 0     
    detectorz   =1036.8
    d=tolerance # border tolerance
    if (0+d) < x < (detectorx-d):
            if (-detectory+d)< y < (detectory-d):
                    if (0+d) < z < (detectorz-d):
                        return True
    return False

def CheckBorderFid(x,y,z,arr):
    detectorx   =256.35     # In cm
    detectory   =116.5      # Symmetric around 0     
    detectorz   =1036.8
    if (0+arr[0][0]) < x < (detectorx-arr[0][1]):
            if (-detectory+arr[1][0])< y < (detectory-arr[1][1]):
                    if (0+arr[2][0]) < z < (detectorz-arr[2][1]):
                        return True
    return False


# Return the angle between two numpy vectors.
def anglevec(v1,v2):
    v1_u = v1 / np.linalg.norm(v1)
    v2_u = v2 / np.linalg.norm(v2)
    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))


# Verifies if the event passed the flash precuts.
def flash_precut(row):
    #Flash precuts
    t_start = 3.2
    t_end   = t_start+1.6
    min_PE  = 50

    for time,PE in zip(row['flash_time'],row['flash_PE']):
        if time>t_start and time<t_end and PE>min_PE:
            return True
    return False

# Sum reconstructed energy
def CalcRecoE(row):
    reco_energy = [0,0,0,0]
    total_nhits = [0,0,0]
    
    for tr_hits,tr_dedx,nhits in zip(row['track_energy_hits'],row['track_energy_dedx'],row["track_nhits"]):
        reco_energy[:3]+=tr_hits  
        reco_energy[-1]+=tr_dedx 
        total_nhits+=nhits
        
    for shower,nhits in zip(row['shower_energy'],row["shower_nhits"]):
        reco_energy[:3]+=shower
        reco_energy[-1]+=shower[2] #4th entry uses plane 2 hits for showers and dedx for tracks
        total_nhits+=nhits
        
    return pd.Series({"reconstructed_energy": reco_energy, "total_nhits": total_nhits})

# Definition 
def true_thresholds(row):
    passed_e=False
    passed_p=False
    for pdg,E in zip(row["nu_daughters_pdg"],row["nu_daughters_E"]):
        if pdg==11 and E>min_e:
            passed_e=True
        if pdg==2212 and E>min_p:
            passed_p=True
    return passed_e and passed_p

# Formatting
def sciNot(x):
    x=float(x)
    return "{:.1f}".format(x)

# Calculates the true end point for electron showers, for 95% of energy
def showerTrueEnd(row):
    true_shower_end=[0,0,0]
    
    i_daughter = np.argwhere(row['nu_daughters_pdg']==11)
    i_shower   = np.argwhere(row['true_shower_pdg']==11)
    if len(i_daughter[0])>1:
        print('More than 1 true electron daughter')
    if len(i_shower[0])  >1:
        print('More than 1 true electron shower')
    i_daughter = i_daughter[0][0]
    i_shower   = i_shower[0][0]
    
    E_ratio = (row['nu_daughters_E'].values)[i_daughter]/E_c
    t_max = np.log(E_ratio)-1.0
    length = (t_max+0.08*18+9.0)*X_o
    
    direction = np.array([row['nu_daughters_px'].values[i_daughter],row['nu_daughters_py'].values[i_daughter],row['nu_daughters_pz'].values[i_daughter]])
    true_shower_start = np.array([row['nu_daughters_vx'].values[i_daughter],row['nu_daughters_vy'].values[i_daughter],row['nu_daughters_vz'].values[i_daughter]])
    true_shower_end = true_shower_start+length*direction/np.linalg.norm(direction)
    
    return pd.Series({"true_shower_endx": true_shower_end[0], "true_shower_endy": true_shower_end[1], "true_shower_endz": true_shower_end[2] })

def showerRecoEnd(row):
    for x,y,z,px,py,pz,l in zip(row['shower_start_x'],row['shower_start_y'],row['shower_start_z'],row['shower_dir_x'],row['shower_dir_y'],row['shower_dir_z'],row['shower_length']):
        start=np.array([x,y,z])
        direc=np.array([px,py,pz])
        end=start+l*direc/np.linalg.norm(direc)
        if not CheckBorderTPC(*end):
            return False
    return True

# Calculates the percentage of sps inside the fiducial volume and the percentage of deposited collection plane charge
def containedPercent(row):
    n1=0
    n2=0
    d1=0.1
    d2=0.1
    
    for x,y,z,q in zip(row['shower_sp_x'],row['shower_sp_y'],row['shower_sp_z'],row['shower_sp_int']):
        d1+=1
        if q>0:
            d2+=q
        if CheckBorderFid(x,y,z,fid_arr):
            n1+=1
            if q>0:
                n2+=q

    return pd.Series({"shower_containment_h": n1/d1, "shower_containment_q": n2/d2 })
```

```python
# Load root files into a dataframe or dictionary of dataframes.
def loadData(filedict,                  # List of input files, dictionary
             ignorelist,                # List of fields (columns) to exclude from the dataframe  
             maxn=4300,                 # Maximum number of events (rows) in the final dataframe 
             concat=False,              # In case of multiple input tags, merge dataframes?
             contained=True,            # Is the true neutrino interaction in a defined volume?
             passed=False,              # Did the event pass the selection stage?
             flashpassed=False,         # Was there still a candidate passing the optical cuts?
             truecontains=[11,2212],    # Require the true neutrino interaction to have the following daughters.
             LEE_denom=False ):         # 1 proton of at least 40MeV, 1 electron of at least 20MeV, true vertex inside: x[10,10],y[20,20],z[10,50]
    
    global categories
    global fid_arr
    dfdict={}
    
    for key,fname in filedict.items():
        dftemp=read_root(fname,"robertoana/pandoratree",ignore=ignorelist)
        
        print ('\n\nDataframe loaded from:',fname)
        print ('Dataframe', key, 'has', len(dftemp.index), 'entries.')
        
        if contained:
            dftemp=dftemp[dftemp.apply(lambda x: CheckBorderTPC(x['true_vx'],x['true_vy'],x['true_vz']), axis=1)]
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries with true vertex in TPC.')
            
        for pdg in truecontains:
            dftemp = dftemp[dftemp["nu_daughters_pdg"].apply(lambda x: pdg in x)]
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries with a pdg',pdg,'particle.')
            
        if LEE_denom:
            dftemp=dftemp[dftemp.apply(lambda x: CheckBorderFid(x['true_vx'],x['true_vy'],x['true_vz'],fid_arr), axis=1)]
            #dftemp.drop(['true_vx','true_vy','true_vz'],axis=1,inplace=True)
            dftemp = dftemp[dftemp[["nu_daughters_pdg","nu_daughters_E"]].apply(true_thresholds, axis=1)]
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries with the LEE signal definition.')
        
        denom = len(dftemp.index)
        
        dftemp = dftemp[dftemp.apply(flash_precut, axis=1)]
        print ('Dataframe', key, 'has', len(dftemp.index), 'entries with a valid flash.')
        
        if(maxn>0 and maxn<len(dftemp.index)):
            dftemp=dftemp.sample(n=maxn)
            print ('Final dataframe', key, 'has', len(dftemp.index), 'entries.') 

            
        if flashpassed:
            dftemp = dftemp[ dftemp['flash_passed'].apply(lambda x: not np.all(x==-1))]
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries passing the optical selection.')
            
        if passed:
            dftemp=dftemp[dftemp['passed']==1]
            total = len(dftemp.index)
            print ('Dataframe', key, 'has', len(dftemp.index), 'entries which passed selection.')
            #Roberto Efficiency
            r_eff = len(dftemp[dftemp['category'] !=1 ].index)
            print ('Category efficiency :', r_eff,'/', denom,'(',sciNot(r_eff/denom*100),'%)','signal events passed and was not a cosmic.')
            
            #Roxanne Purity
            close = len(dftemp[dftemp['distance']<5].index)
            print ('Closeness purity :', close,'/', total,'(',sciNot(close/total*100),'%)','of passed events is within 5cm from true_sce vertex.')
            
    dftemp["category"]=dftemp["category"].map(categories)        
    dftemp.drop(['reconstructed_energy'],axis=1,inplace=True) #because there was a bug
    dftemp[['reconstructed_energy','total_nhits']]=dftemp[["track_energy_hits","shower_energy",'track_energy_dedx',"shower_nhits","track_nhits"]].apply(CalcRecoE,axis=1)
    
    dftemp["flash_config"]=key
    dfdict[key]=dftemp
        
    if concat:
        return pd.concat(dfdict.values(),ignore_index=True,copy=False)
    else:
        return dfdict
```

```python
# Shower containment
plt.style.use('default')
fig,ax=plt.subplots(ncols=2,nrows=1,figsize=(4*gr,4),sharey=True)
plt.suptitle(r'Shower containment')
N=50

arr_cont_h  = df_shower[df_shower['e_contained']==1]['shower_containment_h'].values
arr_cont_q  = df_shower[df_shower['e_contained']==1]['shower_containment_q'].values
arr_ncont_h = df_shower[df_shower['e_contained']==0]['shower_containment_h'].values
arr_ncont_q = df_shower[df_shower['e_contained']==0]['shower_containment_q'].values
labels=('True contained','True not contained')

ax[0].set_title('Using number of spacepoints')
ax[0].hist([arr_cont_h,arr_ncont_h],histtype='step', bins=25, fill=False,label=labels)
ax[0].set_xlabel('containment [%]')
ax[0].set_ylabel('#Events')
ax[0].legend(loc=2)
ax[1].set_title('Using collection plane charge')
ax[1].hist([arr_cont_q,arr_ncont_q],histtype='step', bins=25, fill=False,label=labels)
ax[1].set_xlabel('containment [%]')
ax[1].legend(loc=2)

plt.tight_layout()
fig.subplots_adjust(top=0.85)
plt.savefig('shower_sp_cont.pdf')

pass_q=0
stop_q=0
pass_length = sum(df_shower[df_shower['e_contained']==1]['reco_length_containment'].values)
stop_length = sum(df_shower[df_shower['e_contained']==0]['reco_length_containment'].values)

for cont_h,cont_q in zip(arr_cont_h,arr_cont_q):
    if cont_h>.8 and cont_q>.8:
        pass_q+=1
        
for ncont_h,ncont_q in zip(arr_ncont_h,arr_ncont_q):
    if ncont_h>.8 and ncont_q>.8:
        stop_q+=1
        
print('SP: Passing rate true contained: '         ,pass_q/len(arr_cont_h)*100)
print('SP: Passing rate true not-contained: '     ,stop_q/len(arr_ncont_h)*100)
print('LENGTH: Passing rate true contained: '     ,pass_length/len(arr_cont_h)*100)
print('LENGTH: Passing rate true not-contained: ' ,stop_length/len(arr_ncont_h)*100)
```
